## Step 10
Explore the smaller data frame:

```
# look at the first 10 colleges
df_smaller.head(10)
```

```
# look at the last 10 colleges
df_smaller.tail(10)
```

```
# look at a random sample of 10 colleges
df_smaller.sample(10)
```

```
# Find and display UMBC
df_bc = df_smaller.loc[df_smaller['INSTNM'] == 'University of Maryland-Baltimore County']
df_bc
```

```
# Find and display Univ. of Maryland College Park
df_cp = df_smaller.loc[df_smaller['INSTNM'] == 'University of Maryland-College Park']
df_cp
```

```
# Find and display Loyola College Maryland
df_l = df_smaller.loc[df_smaller['INSTNM'] == 'Loyola University Maryland']
df_l
```

```
# Find and display Johns Hopkins
df_jh = df_smaller.loc[df_smaller['INSTNM'] == 'Johns Hopkins University']
df_jh
```

```
# Create a dataframe with the above four instiutions
df_MD = pd.concat([df_bc,df_cp,df_l,df_jh], axis=0)
df_MD.head()
```

```
# Calculate the min
print("Min of TUITIONFEE_IN:", df_MD['TUITIONFEE_IN'].min())

# Calculate the max
print("Max of TUITIONFEE_IN:", df_MD['TUITIONFEE_IN'].max())

# Calculate the average
print("Average of TUITIONFEE_IN:", df_MD['TUITIONFEE_IN'].mean())
```

## Step 11
Filter the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia. Save the filtered dataset to a new file named "Five_States_Colleges.csv" for future analysis. Upload this file to your GitHub along with the notebook.
